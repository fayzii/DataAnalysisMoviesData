```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('top.csv',usecols=['release_year_top','popularity_top','genres_top'])
data.head()
```

```python
data.dropna(how='any',inplace=True)
data.isnull().sum()
```

```python
df=pd.DataFrame(data)
df.head()
```

```python
newGenres=df.genres_top.str.split('|',expand=True).stack()
df.join(pd.Series(index=newGenres.index.droplevel(1),data=newGenres.values,name='modified_genres'))
```

```python
df.to_csv('splittedGenres.csv')
```
